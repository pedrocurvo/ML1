```python
import numpy as np
import matplotlib.pyplot as plt

# Adjust the code to generate diagonal points with alternating colors
np.random.seed(42)  # For reproducibility

# Generate diagonal points alternating red and blue
n_points = 200
x_vals = np.linspace(-10, 10, n_points)
y_vals = x_vals  # Diagonal points

# Assign alternating labels (1 = red, -1 = blue) in intervals of 10 points
labels = np.array([1 if i % 20 < 10 else -1 for i in range(n_points)])

# Apply the transformation: [cos(x1 - x2), sin(x1 - x2)]
cos_vals = np.cos(x_vals)
sin_vals = np.sin(y_vals)

# Plot the diagonal points
plt.figure(figsize=(12, 6))

# Original diagonal pattern
plt.subplot(1, 2, 1)
for i in range(n_points):
    color = 'red' if labels[i] == 1 else 'blue'
    plt.scatter(x_vals[i], y_vals[i], color=color)
plt.title('Original Diagonal Points with Alternating Colors')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

# Transformed points using cos and sin
plt.subplot(1, 2, 2)
for i in range(n_points):
    color = 'red' if labels[i] == 1 else 'blue'
    plt.scatter(cos_vals[i], sin_vals[i], color=color)
plt.title('After Applying $\\cos(x_1 - x_2)$ and $\\sin(x_1 - x_2)$')
plt.xlabel('$\\cos(x_1 - x_2)$')
plt.ylabel('$\\sin(x_1 - x_2)$')

# Draw decision boundary w_1 * cos(x1 - x2) + w_2 * sin(x1 - x2) + b = 0
w_1 = 2
w_2 = 2
b = 1
yb_vals = (- w_1 * cos_vals - b) / w_2
plt.plot(cos_vals, yb_vals, color='black')

plt.tight_layout()
plt.show()
```

```python
# Generate sin x/x data points if >0 red, else blue
n_points = 200
x_vals = np.linspace(-10, 10, n_points)
y_vals = np.sin(x_vals) / x_vals

# Assign labels (1 = red, -1 = blue) based on the sign of sin(x)/x
labels = np.array([1 if np.sin(x) / x > 0 else -1 for x in x_vals])

# Draw the sin x/x data points

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
for i in range(n_points):
    color = 'red' if labels[i] == 1 else 'blue'
    plt.scatter(x_vals[i], y_vals[i], color=color)
plt.xlabel('$x$')
plt.ylabel('$\\frac{\\sin(x)}{x}$')
plt.tight_layout()
plt.savefig('sin_x_over_x.png')
plt.show()
```
